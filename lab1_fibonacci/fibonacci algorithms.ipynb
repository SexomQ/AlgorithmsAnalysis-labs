```python
from math import sqrt, pow
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import time as tm
import pandas as pd
import pprint as pp
import decimal as dec
```

```python
# styles
sns.set_style("darkgrid")

#initialization
terms_rec = [5, 7, 10, 12, 15, 17, 20, 22, 25, 27, 30, 32, 35, 37]
terms = [501, 631, 794, 1000, 1259, 1585, 1995, 2512, 3162, 3981, 5012, 6310, 7943, 10000, 12589, 15849]
```

```python
# recursion
def fibonacci_rec(no):
    if(no == 0):
        return 0
    elif(no == 1):
        return 1
    else:
        return (fibonacci_rec(no - 1) + fibonacci_rec(no - 2))

values = []
exec_time = []

for i in terms_rec:
    start_time = tm.time()
    values.append(fibonacci_rec(i))
    exec_time.append(tm.time() - start_time)

plt.plot(terms_rec, exec_time)
plt.xlabel("n-th term")
plt.ylabel("time (seconds)")
plt.show()

# print(values)
# print(exec_time)

table = pd.DataFrame(data=terms_rec, columns=['term'])
table['value'] = values
table['time (seconds)'] = exec_time
print(table)
```

```python
# dynamic
def fibonacci_dynamic(n):
    f = [0, 1]
    for i in range(2, n+1):
        f.append(f[i-1] + f[i-2])
    return f[-1]

values = []
exec_time = []

for i in terms:
    start_time = tm.time()
    values.append(fibonacci_dynamic(i))
    exec_time.append(tm.time() - start_time)

plt.plot(terms, exec_time)
plt.xlabel("n-th term")
plt.ylabel("time (seconds)")
plt.show()

# print(values)
# print(exec_time)

table = pd.DataFrame(data=terms, columns=['term'])
table['value'] = values
table['time (seconds)'] = exec_time
print(table)
```

```python
# Binet formula
def fibonacci_formula(n):
    dec.getcontext().prec = 10000
    sqrt_5 = dec.Decimal(5).sqrt()
    
    phi = ((1 + sqrt_5) / 2)
    val = ((phi ** n) - ((-phi) ** -n)) / sqrt_5

    return round(val)

values = []
exec_time = []

for i in terms:
    start_time = tm.time()
    values.append(fibonacci_formula(i))
    exec_time.append(tm.time() - start_time)

plt.plot(terms, exec_time)
plt.xlabel("n-th term")
plt.ylabel("time (seconds)")
plt.show()

# print(values)
# print(exec_time)

table = pd.DataFrame(data=terms, columns=['term'])
table['value'] = values
table['time (seconds)'] = exec_time
print(table)
```

```python
# iteration
def fibonacci_iter(n):
    val1, val2 = 0, 1

    for i in range(n):
        val1, val2 = val2, val1 + val2

    return val1

values = []
exec_time = []

for i in terms:
    start_time = tm.time()
    values.append(fibonacci_formula(i))
    exec_time.append(tm.time() - start_time)

plt.plot(terms, exec_time)
plt.xlabel("n-th term")
plt.ylabel("time (seconds)")
plt.show()

# print(values)
# print(exec_time)

table = pd.DataFrame(data=terms, columns=['term'])
table['value'] = values
table['time (seconds)'] = exec_time
print(table)
```

```python
# matrix exponentiation   fib(n) = ([[1, 1][1, 0]] ** n) [0,1]
def fibonacci_matrix(n): 
    init = np.array([[1,1],[1,0]], dtype='object') 
    value_matrix = np.linalg.matrix_power(init, n)
    return value_matrix[0,1]

values = []
exec_time = []

for i in terms:
    start_time = tm.time()
    values.append(fibonacci_matrix(i))
    exec_time.append(tm.time() - start_time)

plt.plot(terms, exec_time)
plt.xlabel("n-th term")
plt.ylabel("time (seconds)")
plt.show()

# print(values)
# print(exec_time)

table = pd.DataFrame(data=terms, columns=['term'])
table['value'] = values
table['time (seconds)'] = exec_time
print(table)
```

```python
# iterative doubling method
def fibonacci_doubling(n):
    bin_of_n = bin(n)[2:]  # binary string of n
    f = [0, 1]  # [F(i), F(i+1)] => i=0
    
    for b in bin_of_n:
        f2i1 = f[1]**2 + f[0]**2  # F(2i+1)
        f2i = f[0]*(2*f[1]-f[0])  # F(2i)
    
        if b == '0':
            f[0], f[1] = f2i, f2i1  # [F(2i), F(2i+1)]
        else:
            f[0], f[1] = f2i1, f2i1+f2i  # [F(2i+1), F(2i+2)]
    return f[0]

values = []
exec_time = []

for i in terms:
    start_time = tm.time()
    values.append(fibonacci_doubling(i))
    exec_time.append(tm.time() - start_time)

plt.plot(terms, exec_time)
plt.xlabel("n-th term")
plt.ylabel("time (seconds)")
plt.show()

# print(values)
# print(exec_time)

table = pd.DataFrame(data=terms, columns=['term'])
table['value'] = values
table['time (seconds)'] = exec_time
print(table)
```
