### 1. Recursion method

```python
def fib(n):
    if n <= 1:
        return n
    return fib(n-1) + fib(n-2)
```

```python
rec_list = [5, 7, 10, 12, 15, 17, 20, 22, 25, 27, 30, 32, 35, 37, 40, 42, 45]
```

```python
ex_time_list = []
```

```python
table1 = pd.DataFrame(columns=['Fibonacci term number', 'Fibonacci number', 'Execution time(wall time)', 'Execution time(CPU time)'])
```

```python
for i in rec_list:
    """
    loop for calculating the fibonacci number
    and execution time and including the results
    in a table and an array
    """
    start = time.time()
    start2 = time.process_time()
    nr = fib(i)
    end = time.time()
    end2 = time.process_time()
    # time in milliseconds
    ex_time = (end - start) * 1000
    ex_time_list.append(ex_time)
    ex_time2 = (end2 - start2) * 1000
    new_row = {'Fibonacci term number' : f'{i}', 'Fibonacci number' : f'{nr}', 'Execution time(wall time)' : ex_time, 'Execution time(CPU time)': ex_time2}
    table1 = table1.append(new_row, ignore_index=True)
```

```python
table1
```

#### Let's visualize

```python
plt.plot(rec_list,ex_time_list, '-ob')
plt.xlabel('n-th Fibonacci term')
plt.ylabel('Execution time in milliseconds')
plt.title('Recursion')
plt.show()
```

### 2. Dynamic programming

```python
def fib_dynamic(n):
    l = [0, 1]

    for j in range(2, n + 1):
        l.append(l[j - 1] + l[j - 2])
    return l[n]
```

## Optimized Dynamic Programming

```python
def fib_opt_dynamic(n):
    a = 0
    b = 1
    if n < 0:
        print("Incorrect input")
    elif n == 0:
        return a
    elif n == 1:
        return b
    else:
        for k in range(2,n+1):
            c = a + b
            a = b
            b = c
        return b
```

## Matrix power method

```python
def multiply(F, M):

    x = (F[0][0] * M[0][0] +
         F[0][1] * M[1][0])
    y = (F[0][0] * M[0][1] +
         F[0][1] * M[1][1])
    z = (F[1][0] * M[0][0] +
         F[1][1] * M[1][0])
    w = (F[1][0] * M[0][1] +
         F[1][1] * M[1][1])

    F[0][0] = x
    F[0][1] = y
    F[1][0] = z
    F[1][1] = w

def power(F, n):

    M = [[1, 1],
         [1, 0]]

    for l in range(2, n + 1):
        multiply(F, M)

def fib_mat(n):
    F = [[1, 1],
         [1, 0]]
    if n == 0:
        return 0
    power(F, n - 1)

    return F[0][0]
```

## O(log n) time method

```python
MAX = 1000

# Create an array for memoization
f = [0] * MAX

# Returns n-th fibonacci number using table f[]
def fib(n) :
    # Base cases
    if n == 0:
        return 0
    if n == 1 or n == 2:
        f[n] = 1
        return f[n]

    # If fib(n) is already computed
    if f[n]:
        return f[n]

    if n & 1 :
        k = (n + 1) // 2
    else :
        k = n // 2

    # Applying above formula Note value n&1 is 1
    # if n is odd, else 0.
    if n & 1:
        f[n] = (fib(k) * fib(k) + fib(k-1) * fib(k-1))
    else :
        f[n] = (2*fib(k-1) + fib(k))*fib(k)

    return f[n]
```

## Dynamic Programming using memoization

```python
# Initialize array of dp
dp = [-1 for i in range(10)]

def fib_(n):
    if n <= 1:
        return n
    global dp

    if dp[n - 1] != -1:
        first = dp[n - 1]
    else:
        first = fib(n - 1)
    if dp[n - 2] != -1:
        second = dp[n - 2]
    else:
        second = fib(n - 2)
    dp[n] = first + second

    # Memoization
    return dp[n]
```
