```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import time as tm
import pandas as pd
```

```python
# styles
sns.set_style("darkgrid")
```

```python
check_points = [500, 1500, 5000, 15000, 50000, 150000, 500000, 1000000]
```

```python
# Algorithm 1
def alg_1(n):
    C = [True] * (n+1)
    C[1] = False
    i = 2
    while i <= n:
        if C[i]:
            j = 2 * i
            while j <= n:
                C[j] = False
                j = j + i
        i = i + 1
    return [ind for ind, val in enumerate(C) if C[ind] == True]
```

```python
exec_time_1 = []
for value in check_points:
    start_time = tm.time()
    alg_1(value)
    exec_time_1.append(tm.time() - start_time)

plt.plot(check_points, exec_time_1)
plt.xlabel("array length")
plt.ylabel("time (seconds)")
plt.show()

table = pd.DataFrame(data=check_points, columns=['length'])
table['time (seconds)'] = exec_time_1
print(table)
```

```python
# Algorithm 2
def alg_2(n):
    C = [True] * (n+1)
    C[1] = False
    i = 2
    while i <= n:
        j = 2 * i
        while j <= n:
            C[j] = False
            j = j + i
        i = i + 1
        
    return [ind for ind, val in enumerate(C) if C[ind] == True]
```

```python
exec_time_2 = []
for value in check_points:
    start_time = tm.time()
    alg_2(value)
    exec_time_2.append(tm.time() - start_time)

plt.plot(check_points, exec_time_2)
plt.xlabel("array length")
plt.ylabel("time (seconds)")
plt.show()

table = pd.DataFrame(data=check_points, columns=['length'])
table['time (seconds)'] = exec_time_2
print(table)
```

```python
# Algorithm 3
def alg_3(n):
    C = [True] * (n+1)
    C[1] = False
    i = 2
    while i <= n:
        if C[i]:
            j = i + 1
            while j <= n:
                if j % i == 0:
                    C[j] = False
                j = j + 1
        i = i + 1
    
    return [ind for ind, val in enumerate(C) if C[ind] == True]
```

```python
exec_time_3 = []
for value in check_points[:5]:
    start_time = tm.time()
    alg_3(value)
    exec_time_3.append(tm.time() - start_time)

plt.plot(check_points[:5], exec_time_3)
plt.xlabel("array length")
plt.ylabel("time (seconds)")
plt.show()

table = pd.DataFrame(data=check_points[:5], columns=['length'])
table['time (seconds)'] = exec_time_3
print(table)
```

```python
# Algorithm 4
def alg_4(n):
    C = [True] * (n+1)
    C[1] = False
    i = 2
    while i <= n:
        j = 2
        while j < i:
            if i % j == 0:
                C[i] = False
                break
            j += 1
        i += 1
    return [ind for ind, val in enumerate(C) if C[ind] == True]
```

```python
exec_time_4 = []
for value in check_points[:5]:
    start_time = tm.time()
    alg_4(value)
    exec_time_4.append(tm.time() - start_time)

plt.plot(check_points[:5], exec_time_4)
plt.xlabel("array length")
plt.ylabel("time (seconds)")
plt.show()

table = pd.DataFrame(data=check_points[:5], columns=['length'])
table['time (seconds)'] = exec_time_4
print(table)
```

```python
# Algorithm 5
def alg_5(n):
    C = [True] * (n+1)
    C[1], C[0] = False, False
    i = 2
    while i <= n:
        j = 2
        while j <= i**(1/2):
            if i % j == 0:
                C[i] = False
                break
            j = j + 1
        i = i + 1
    return [ind for ind, val in enumerate(C) if C[ind] == True]
```

```python
exec_time_5 = []
for value in check_points:
    start_time = tm.time()
    alg_5(value)
    exec_time_5.append(tm.time() - start_time)

plt.plot(check_points, exec_time_5)
plt.xlabel("array length")
plt.ylabel("time (seconds)")
plt.show()

table = pd.DataFrame(data=check_points, columns=['length'])
table['time (seconds)'] = exec_time_5
print(table)
```
