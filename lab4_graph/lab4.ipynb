### Implementation of the algorithms (DFS and BFS)

```python
# The DFS algorithm
def dfs(st_node, fin_node, tree):
    visited = set()  # keep track of visited nodes

    def dfs_alg(node, end_node):
        visited.add(node)
        # print(node, end=' ')
        if node == end_node:
            return True

        for neighbor in tree[node]:
            if neighbor not in visited:
                if dfs_alg(neighbor, end_node):
                    return True

        return False

    if dfs_alg(st_node, fin_node):
        return True
    else: return False

# The BFS algorithm
def bfs(st_node, fin_node, tree):
    visited = set()  # keep track of visited nodes
    queue = deque([st_node])  # initialize queue with start node

    while queue:
        node = queue.popleft()
        # print(node, end=' ')
        if node == fin_node:
            return True

        visited.add(node)
        for neighbor in tree[node]:
            if neighbor not in visited:
                queue.append(neighbor)

    return False
```

### Get nodes to find for dfs and bfs

```python
find_nodes = r.sample(nodes, 10)
```

### Run dfs and bfs on the balanced tree

```python
exec_time_dfs_bt = []
no_of_nodes = range(1, len(find_nodes)+1)

print(find_nodes)
for i in range(len(find_nodes)):
    start_time = tm.time()
    tm.sleep(0.000000001)
    for node in find_nodes[:i+1]:

        dfs('S', node, balanced_tree)
    exec_time_dfs_bt.append(tm.time() - start_time)

print(exec_time_dfs_bt)
print(no_of_nodes)
    
plt.plot(no_of_nodes, exec_time_dfs_bt)
plt.xlabel("no of nodes to find")
plt.ylabel("time (seconds)")
plt.show()

table = pd.DataFrame(data=no_of_nodes, columns=['No_nodes'])
table['time (seconds)'] = exec_time_dfs_bt
print(table)
```
